```python
import csv
lst=list()
d=dict()
with open('bank-full.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=';')
    for row in csv_reader:
        p=row[0]
        if p not in d:d[p]=1
        else:
            d[p]+=1
    for key,val in list(d.items()):
        lst.append((val,key))
lst.sort(reverse=True)
print(lst)

        
```

```python

import csv
import matplotlib.pylab as plt
d=dict()
with open('bank-full.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=';')
    for row in csv_reader:
        p=row[1]
        if p not in d:d[p]=1
        else:
            d[p]+=1       
    print(d)
    lst=d.items()
    x,y=zip(*lst)
    plt.bar(x,y)
    plt.xticks(rotation=45, ha='right')
    plt.show()
    
```

```python
count=0
count1=0
c=0
with open('bank-full.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=';')
    for row in csv_reader:
        p=row[5]
        if p=='no':count+=1
        elif p=='yes':
            count1+=1
        else:
            c=c+1

    print("Total number of people with no housing loan ",count)
    print("Total number of people with housing loan ",count1)
    print("Total number of people unknown whether they have housing loan or not ",c)
        
        
```
